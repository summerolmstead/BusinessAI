**Task 1**: Variables and Basic Operations
Create Variables: Define two variables, product_cost and sale_price, and assign them values of your choice.
Calculate Profit: Calculate the profit by subtracting product_cost from sale_price. Store the result in a variable named profit.
Print Results: Print the values of product_cost, sale_price, and profit with a descriptive message.

```python
product_cost = 10
sale_price = 20
profit = sale_price - product_cost
print("The profit from selling a jacket, calculated by subtracting the product cost from the sale price, is "+ str(profit)+".") # lets just assume and put a jacket for the product lol
```

**Task 2**: Control Flow - Making Business Decisions
Discount Logic: Create a variable quantity_sold and assign it a value. Write an if-elif-else structure that applies a 10% discount if more than 50 items are sold, a 5% discount if 20 to 50 items are sold, and no discount otherwise.
Calculate Discounted Price: Calculate the final price after the discount and print the result with a descriptive message.

```python
quantity_sold = 70

if quantity_sold > 50:
    discount = 0.1
elif quantity_sold >= 20:
    discount = 0.05
else:
    discount = 0

final_price = sale_price * (1 - discount)
print("The final price after the discount is "+ str(final_price)+".") #recall it takes prev result from task 1 cell in the sale_price variable declared earlier.
#result for 20$ jacket is 10% off makes sense for discount to now result in 18$ sale price!
```

**Task 3**: Functions - Modularizing the Discount Strategy
Define a Function for Discount: Define a function calculate_discount(price, quantity) that takes the price and quantity as arguments and returns the final price after applying the discount logic from Task 2.
Use the Function: Call the function with different price and quantity values and print the results.

```python
def calculate_discount(price, quantity): #same code just adding function and parameters that need to input price and quantity
    if quantity > 50:
        discount = 0.1
    elif quantity >= 20:
        discount = 0.05
    else:
        discount = 0

    final_price = price * (1 - discount)

    return final_price # returns the discount calculation variable inside function. these variables only exist inside of this function so it is not getting confused that we have the same names

#lets call the function now to showcase the same as before which should result in 18$ for the jacket
print("The discount considering the quanity sold being "+str(quantity_sold)+" and the original price being "+str(sale_price)+" makes the discounted price to be "+str(calculate_discount(20, 70))+".")
```

```python
#hmm lol this is interesting. I will make a randomized data set maybe and then perform some random data analytics on it haha
#importing libraries for data analytics
import numpy as np
import pandas as pd

#number of rows and columns for the dataset
num_rows = 40
num_columns = 18
random_data = [] #initalizing array to put it in

#for loop to just fill it all with random values - note: when this runs it will make a new random dataset everytime - never the same result
for _ in range(num_rows):
    row = [np.random.randint(0, 101) for _ in range(num_columns)]
    random_data.append(row)

#need column names bc rows i assume just are zero based index is fine
columns = [f'Col{i+1}' for i in range(num_columns)]

#creating dataframe for our randomzied dataset created to perform random stats on it
df = pd.DataFrame(random_data, columns=columns)

print(df.head())
```

```python
# so now we had our own randomized dataset! lets see some summary stats
#mean of each column
mean_values = df.mean()
#standard deviation of each column
std_values = df.std()
#summary statistics for df
summary_stats = df.describe()
#printing results
print("Mean values:\n", mean_values)
print("Standard deviation values:\n", std_values)
print("Summary statistics:\n", summary_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#boxplot for each column
sns.boxplot(data=df)
plt.title('Boxplot of Columns')
plt.show()

#scatter plot to see relationships between columns
sns.pairplot(df)
plt.suptitle('Pairplot of Columns', y=1.02) # the y parameter is just for spacing
plt.show()
```
