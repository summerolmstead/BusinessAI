1. Lists and Tuples:
   - Create a list of at least 10 numbers. Then, perform the following operations:
     - Append a new number to the list.
     - Remove a number from the list.
     - Sort the list in ascending order.
     - Convert this list into a tuple.
   - Write a function that takes this tuple as an argument and returns the sum and average of the numbers.

```python
#making a list of 10 numbers
numbers = [10,2,3,5,6,7,4,8,9,1] # initalizing
print(numbers)
#appending a new number to the list
numbers.append(11)
print(numbers)
#removing a number from the list
numbers.remove(5)
print(numbers)
#sorting the list in ascending order
numbers.sort()
print(numbers)
#converting list to tuple data type
numbers_tuple = tuple(numbers)
print(numbers_tuple)

#function that takes tuple as an argument and returns the sum and average of numbers
def sum_avg(numbers_tuple):
    total = sum(numbers_tuple)
    average = total / len(numbers_tuple)
    return print("The total is "+str(total)+". The average is "+str(average)+".")

sum_avg(numbers_tuple)
```

3. Sets:
   - Create two sets of numbers, `set_a` and `set_b`, with at least 5 elements each, some of which should overlap between the sets.
   - Demonstrate the following set operations:
     - Union
     - Intersection
     - Difference (`set_a` - `set_b`)
     - Symmetric Difference

```python
set_a = {1,2,3,4,5}
set_b = {4,5,6,7,8}

#union combining the two sets into one
set_c = set_a.union(set_b)
print(set_c)
#intersection of the two sets
set_d = set_a.intersection(set_b)
print(set_d)
#difference
print(set_a - set_b)
#symmetric difference
print(set_a ^ set_b)
```

1. Numpy:
   - Create a numpy array of shape (5,5) with numbers from 1 to 25.
   - Calculate the transpose of the array.
   - Find the inverse of the array (ensure the array is invertible).


```python
import numpy as np

#array with random integers in the range 1 to 25 and reshaping to (5, 5)
exampleArray = np.random.randint(1, 26, size=(5, 5))
print(exampleArray)
#transpose of the array
transposeArray = np.transpose(exampleArray)
print(transposeArray)
#inverse of the array (if possible if determinant is zero )
if np.linalg.det(exampleArray) != 0:
    inverseArray = np.linalg.inv(exampleArray)
    print("\nInverse of the Array:\n", inverseArray)
else:
    print("\nMatrix is not invertible")
```

2. Pandas:
   - Create a pandas DataFrame with the following columns: `ID`, `Name`, `Age`, and `Department`. Populate it with data for at least 5 individuals.
   - Perform basic data analysis: find the average age and count the number of unique departments.

```python
import pandas as pd

#defining dataframe with id , name , age and department
df = pd.DataFrame(
    {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Anna','Tim','Jane','Kendra','James'],
    'Age': [18,20,24,23,24],
    'Department':['Economics','Finance','Engineering','Computer Science','Economics']
    })

# basic analysis to find average age in df and count number of unique departments
average_age = df['Age'].mean()
unique_departments = df['Department'].nunique()
print("The Average Age is: ", average_age)
print("The Number of Unique Departments is: ", unique_departments)
```

3. OpenCV
   - Read an image using OpenCV (you may use any image of your choice).
   - Convert the image to grayscale and save it with a new name.
   - Display the original and grayscale images side by side using `matplotlib`

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("University_of_Tennessee_at_Chattanooga_athletics_logo.png")  # im just uploading this to my google colab workspace i can add this to the submission of the assignment too

#convert the image to grayscale using open cv library
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#writing to file to save the grayscale image with a new name
cv2.imwrite('grayscale_UTC_logo.jpg', gray_image)

#BGR to RGB for correct color display with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#plotting both images side by side
plt.figure(figsize=(10, 5))

#display original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')  #hiding axis dont need to show it

#display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show() #cool it works and looks nice honestly!
```
